Text Cleaning (Lowercase, Trim, Remove Punctuation)

```python
# Text cleaning that preserves NaN and does NOT touch genre columns
import re, string, pandas as pd

def clean_text(s):
    if pd.isna(s):          # keep true NaN
        return pd.NA
    s = str(s).lower().strip()
    s = s.translate(str.maketrans('', '', string.punctuation))
    s = re.sub(r"[^\w\s]", " ", s)   # remove emojis/specials
    s = re.sub(r"\s+", " ", s)
    return s.strip()

text_cols = ['track_name','track_artist','track_album_name','playlist_name']  # exclude genres
df[text_cols] = df[text_cols].apply(lambda col: col.map(clean_text))
```

Date column inspection

```python
import re

year_only = df['track_album_release_date'].astype(str).str.fullmatch(r"\d{4}").sum()
year_month = df['track_album_release_date'].astype(str).str.fullmatch(r"\d{4}-\d{2}").sum()
full_date = df['track_album_release_date'].astype(str).str.fullmatch(r"\d{4}-\d{2}-\d{2}").sum()
invalid = len(df) - (year_only + year_month + full_date)

print(f"Year only: {year_only}")
print(f"Year and month: {year_month}")
print(f"Full date: {full_date}")
print(f"Invalid / missing format: {invalid}")
```

Convert object columns to category

```python
text_cols = ['track_name', 'track_artist', 'track_album_name', 'playlist_name']
df[text_cols] = df[text_cols].astype('string')

# Convert only genre columns to category
genre_cols = ['playlist_genre', 'playlist_subgenre']
df[genre_cols] = df[genre_cols].astype('category')

# Quick check
df[genre_cols].info()
```

Check for duplicates

```python
# Count full-row duplicates
full_dupes = df[df.duplicated(keep=False)]
print(f"Full-row duplicates found: {full_dupes.shape[0]}")

# Display a sample of full-row duplicates (if any)
if not full_dupes.empty:
    print("\nSample of full-row duplicates:")
    display(full_dupes.head())

# Count duplicates by track_id (same song id)
id_dupes = df[df.duplicated(subset=['track_id'], keep=False)]
print(f"\nDuplicates by track_id found: {id_dupes.shape[0]}")

# Display a sample of track_id duplicates (if any)
if not id_dupes.empty:
    print("\nSample of track_id duplicates:")
    display(id_dupes.head())
```
